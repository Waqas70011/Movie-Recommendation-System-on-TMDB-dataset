```python
# import libraries
import numpy as np 
import pandas as pd 
import ast
```

```python
# importing datasets 
movies = pd.read_csv('./datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('./datasets/tmdb_5000_credits.csv')
```

```python
# merging both datasets into a single dataset and reassign it to movies
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

# List of columns to keep

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
# All the needed columns are extracted from movies dataset and assigned to the same `movies`
movies.head(1)
```

```python
# checking for missing data
movies.isnull().sum()
```

```python
# Since we have small number of null values we can drop them.
movies.dropna(subset=['overview'], inplace=True)
```

```python
# rechecking for the results.
movies.isnull().sum()
```

```python
# function for cleaning genres and keywords columns and replacing it with the respective columns for all records.

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# function for selecting first 3 actor names from cast and replacing it with cast column for all records.

def convert_c(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter +=1
        else:
            break
    return l

movies['cast'] = movies['cast'].apply(convert_c)
```

```python
movies['cast'] = movies['cast'].apply(convert_c)
```

```python
# function for selecting only Director name from the crew and then replacing it with crew column for all records.

def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# Overview of the dataset for inspection
movies.head()
```

```python
# converting the overview column data into a list
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda xx: [i.replace(' ','') for i in xx])
movies['keywords'] = movies['keywords'].apply(lambda xx: [i.replace(' ','') for i in xx])
movies['crew'] = movies['crew'].apply(lambda xx: [i.replace(' ','') for i in xx])
movies['cast'] = movies['cast'].apply(lambda xx: [i.replace(' ','') for i in xx])
```

```python
movies.head()
```

```python
# Now concatenating... columns to create a new column called tags

movies['tags'] = movies['overview'].astype(str) + ' ' + movies['keywords'].astype(str) + ' ' + movies['genres'].astype(str) + ' ' + movies['cast'].astype(str) + ' ' + movies['crew'].astype(str)
```

```python
# Now concatenating... columns to create a new column called tags

movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```
